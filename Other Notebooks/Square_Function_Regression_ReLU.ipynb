```python
%matplotlib inline

# import tensorflow
import tensorflow as tf
tf.logging.set_verbosity(tf.logging.ERROR)

# import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt

# define number of neurons in the first layer
NUMBER_OF_NEURONS_IN_FIRST_LAYER = 3

# setup training data
input_x = list(np.arange(0, 20, 0.5))
output_square_x = [np.square(x) for x in input_x]

# plot square(x)
#xvalues_square = np.linspace(0, 20)
#yvalues_square = np.square(xvalues_square)
#plt.plot(xvalues_square, yvalues_square, lw=1, color='red', label='square(x)')

# Auto generate x number value list by arange function.
xvalues_training = list(np.arange(0, 20, 0.2))
    
# Y axis value is assigned training input
yvalues_training = [np.square(x) for x in xvalues_training]

# Draw points based on above x, y value list and remove the point black outline. And the point color is green.
plt.scatter(xvalues_training, yvalues_training, s=5, edgecolors='none', c='red', label='training')

plt.xlabel('x')
plt.ylabel('training values of square(x)')
plt.axhline(0, lw=0.5, color='black')
plt.axvline(0, lw=0.5, color='black')
plt.legend()
plt.show()

# build ml model
prelu = tf.keras.layers.PReLU(alpha_initializer='zeros', alpha_regularizer=None, alpha_constraint=None, shared_axes=None)
l0 = tf.keras.layers.Dense(units=NUMBER_OF_NEURONS_IN_FIRST_LAYER, input_shape=[1], activation=prelu)  
l1 = tf.keras.layers.Dense(units=1, input_shape=[NUMBER_OF_NEURONS_IN_FIRST_LAYER])  
#l2 = tf.keras.layers.Dense(units=1, input_shape=[1], activation=tf.nn.relu)  
model = tf.keras.Sequential([l0, l1])
#model = tf.keras.Sequential([l0])

model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))

print("Number of neurons in the first layer: {}".format(NUMBER_OF_NEURONS_IN_FIRST_LAYER))
print("Start training the model...")
history = model.fit(input_x, output_square_x, epochs=500, verbose=False)
print("Finished training the model.")

# plot ml model with predictions and function square(x)
#xvalues = np.linspace(0, 20)
#yvalues_input = np.square(xvalues)
#plt.plot(xvalues, yvalues_input, lw=1, color='red', label='square(x)')

# Auto generate x number value list by arange function.
xvalues_training = list(np.arange(0, 20, 0.2))
    
# Y axis value is assigned training input
yvalues_training = [np.square(x) for x in xvalues_training]

# Draw points based on above x, y value list and remove the point black outline. And the point color is green.
plt.scatter(xvalues_training, yvalues_training, s=5, edgecolors='none', c='red', label='training')

# Auto generate x number value list by range function.
xvalues_prediction = list(np.arange(0, 20, 0.2))
    
# Y axis value is assigned model predictions
yvalues_output = [model.predict([x]) for x in xvalues_prediction]

# Draw points based on above x, y value list and remove the point black outline. And the point color is green.
plt.scatter(xvalues_prediction, yvalues_output, s=5, edgecolors='none', c='blue', label='prediction')

plt.xlabel('x')
plt.ylabel('training and predicted square(x)')
plt.axhline(0, lw=0.5, color='black')
plt.axvline(0, lw=0.5, color='black')
plt.legend()
plt.show()


```
